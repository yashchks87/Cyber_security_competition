```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
spam = pd.read_csv('./FinalDataset/Spam.csv')
malware = pd.read_csv('./FinalDataset/Malware.csv')
```

```python
frames = [spam, malware]
```

```python
finalData = pd.concat(frames)
```

```python
finalData.shape
```

```python
# With benign
x = finalData['URL_Type_obf_Type'].value_counts()
x = x.to_dict()
keys_final = list(x.keys())
values_final = list(x.values())
plt.xlim(6000, 16000)
plt.barh(keys_final, values_final)
```

```python
# Without benigh
# Malware and spam speelling is capital
# Scale start from 0 and interval as 500
# Scale start from 0 and interval as 1000
y = finalData[finalData.URL_Type_obf_Type != 'benign']
y = y['URL_Type_obf_Type'].value_counts()
y = y.to_dict()
keys_final1 = list(y.keys())
values_final1 = list(y.values())
fig, ax = plt.subplots(figsize=(15, 5))
plt.bar(keys_final1, values_final1)
plt.ylim(0, 7500)
plt.yticks(range(0, 7500, 1200), fontsize= 15)
plt.xticks(fontsize=15)
plt.xlabel("Type of Malicious URL", fontsize=25)
plt.ylabel("Count of Malicious URLs", fontsize=25)
# plt.title("Count of malicious URL", fontsize=30)
# Change orientation...
```
